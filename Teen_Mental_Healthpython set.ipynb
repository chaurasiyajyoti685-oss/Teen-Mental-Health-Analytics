```python
import matplotlib.pyplot as plt

stress_gender = df.groupby('gender')['stress_level'].mean()

stress_gender.plot(kind='bar')

plt.title("Average Stress by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Stress Level")

plt.show()
```

```python
#Histogram - Sleep Hours

import matplotlib.pyplot as plt

plt.hist(df['sleep_hours'], bins=15)

plt.title("Sleep Hours Distribution")
plt.xlabel("Sleep Hours")
plt.ylabel("Number of Students")

plt.show()
```

```python
#Pie Chart - Depression Distribution
depression = df['depression_label'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    depression,
    labels=depression.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Depression Distribution")

plt.show()
```

```python
#Scatter Plot - Sleep vs Stress
plt.figure(figsize=(8,5))

plt.scatter(
    df['sleep_hours'],
    df['stress_level']
)

plt.title("Sleep Hours vs Stress Level")
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")

plt.show()
```

```python
#Bar Chart - Anxiety by Gender

anxiety = df.groupby('gender')['anxiety_level'].mean()

anxiety.plot(kind='bar')

plt.title("Average Anxiety by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Anxiety")

plt.show()
```

```python
#Bar Chart - Social Media Usage
social = df['platform_usage'].value_counts()

social.plot(kind='bar')

plt.title("Most Used Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Users")

plt.show()
```
